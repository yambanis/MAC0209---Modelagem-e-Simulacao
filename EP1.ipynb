```python
import math
import matplotlib.pyplot as pyplot

def nextTmu(x, v):
    """ nextXeuler implementa o método de Euler. 
        
    Parâmetros:
    -----------
    x: float - Posição atual
    t: float - Tempo atual
    v: float - Velocidade do movimento
    
    Returns:
    --------
    x: float : posição atualizada
    
    """
    return x/v
    
def main():
    
    # vetor com as posições onde se encontravam os cronômetros(na forma regular e na forma alternada)
    xOriginal= [0, 10, 20, 30]
    xAlternado= [0, 5, 10, 15, 20, 25, 30]
    
    # vetor para guardar os tempos do movimento
    tOriginalGui=[]
    tAlternadoGui=[]
    
    # velocidade media do Guilherme nas 3 travessias
    v = 0.9264
    for x in xOriginal:
        tOriginalGui.append(nextTmu(x, v))
    for x in xAlternado:
        tAlternadoGui.append(nextTmu(x, v))
        
    pyplot.figure(0)
    pyplot.plot(xOriginal,tOriginalGui,label='Original') 
    pyplot.plot(xAlternado,tAlternadoGui,label='Alternado',linestyle='',marker='o')
    pyplot.title('Time x Space - Guilherme')
    pyplot.show(block=False)

main() 
```
