```python
import math
import matplotlib.pyplot as pyplot

def nextTmu(x, v):
    """ nextXeuler implementa o método de Euler. 
        
    Parâmetros:
    -----------
    x: float - Posição atual
    t: float - Tempo atual
    v: float - Velocidade do movimento
    
    Returns:
    --------
    x: float : posição atualizada
    
    """
    return x/v
    
def main():
    
    # vetor com as posições(m) onde se encontravam os cronômetros na forma Original
    xOriginal= [0, 10, 20, 30]
    
    # vetor para guardar os tempos(s) do movimento
    tAnaliticoGui=[]
    tObtidoGui= [0, 10.405, 20.96, 31.465]
    
    # velocidade media(m/s) do Guilherme nas 3 travessias
    v = 0.9264
    for x in xOriginal:
        tAnaliticoGui.append(nextTmu(x, v))
        
    pyplot.figure(0)
    pyplot.plot(xOriginal,tAnaliticoGui,label='Analítico') 
    pyplot.plot(xOriginal,tObtidoGui,label='Obtido',linestyle='',marker='o')
    pyplot.title('Espaço x Tempo 1- Guilherme')
    pyplot.show(block=False)

main() 
```
