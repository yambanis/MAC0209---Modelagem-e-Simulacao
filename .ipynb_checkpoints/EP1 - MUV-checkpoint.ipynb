```python
import math
import matplotlib.pyplot as pyplot

def nextTmu(x, a):
    """ nextTmu implementa o método analítico para calcular o tempo dado uma velocidade média. 
        
    Parâmetros:
    -----------
    x: float - Posição atual
    v: float - Velocidade média do movimento
    
    Returns:
    --------
    t: float : tempo pra realizar o percurso
    
    """
    return math.sqrt(2*x/a)
    
def main():
    # vetor com as posições onde se encontravam os cronômetros na forma Original
    xOriginal= [0, 10, 20, 30]
    # aceleração média da pessoa 1
    a= 0.5357
    # vetor para guardar os tempos do movimento
    tMUVAnalitico1=[]
    tMUVObtido1= [0, 5.345, 7.885, 10.57]
    xAnalitico=[]
    
    for x in range(0, 30):
        xAnalitico.append(x)
        tMUVAnalitico1.append(nextTmu(x, a))
    
    pyplot.figure(0)
    pyplot.plot(xAnalitico ,tMUVAnalitico1,label='Analitico') 
    pyplot.plot(xOriginal,tMUVObtido1,label='Obtido',linestyle='',marker='o')
    pyplot.title('Espaço x Tempo MUV 1- Pessoa 1')
    pyplot.show(block=False)

main() 
```

```python
import csv
from matplotlib import pyplot as plt

p1 = [0, 10, 20, 30]
p2 = [0, 5, 10, 15, 20, 25, 30]
t = []

with open ("", "r") as mu1:
    rd = csv.reader(mu1, quoting=csv.QUOTE_NONNUMERIC)
    for line in rd:
        t.append([0] + line)
        
        
t_seq = (t[6][1:] + t[7])
t_seq.sort()

plt.figure(figsize=(10,10))
plt.title("Travessia 1.1 - MU: cronômetros pareados")   
plt.xlabel('distância (m)')
plt.ylabel('tempo (s)')
plt.plot(p1, t[0], 'c--', label = "T1: esquerdo")
plt.plot(p1, t[1],'b--', label = "T1: direito")
plt.legend(loc = 0)

plt.figure(figsize=(10,10))
plt.title("Travessia 1.2 - MU: cronômetros pareados")   
plt.xlabel('distância (m)')
plt.ylabel('tempo (s)')
plt.plot(p1, t[3], 'r--', label = "T2: esquerdo")
plt.plot(p1, t[4], 'm--', label = "T2: direito")
plt.legend(loc = 0)

plt.figure(figsize=(10,10))
plt.title("Travessia 1.3 - MU: cronômetros invertidos")   
plt.xlabel('distância (m)')
plt.ylabel('tempo (s)')
plt.plot(p2, t_seq, 'g--', label = "T3: invertidos")
plt.legend(loc = 0)

tSimu = []
# velocidade media do Guilherme nas 3 travessias
v = 0.9264
for x in p2:
    tSimu.append(nextTmu(x, v))

plt.figure(figsize=(10,10))
plt.title("Simulação 1 - MU")   
plt.xlabel('distância (m)')
plt.ylabel('tempo (s)')
plt.plot(p2, tSimu, 'k--', label = "TS: simulação")
plt.legend(loc = 0)

erro = [abs(x - y) for x, y in zip(tSimu, t_seq)]
plt.figure(figsize=(10,10))
plt.title("Erro - travessia 1")   
plt.xlabel('distância (m)')
plt.ylabel('erro absoluto (s)')
plt.plot(p2, erro, 'bs')

```
