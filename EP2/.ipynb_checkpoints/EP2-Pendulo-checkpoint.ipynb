```python
import csv
import pandas
import math
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
def prepara_figura(titulo, eixo_x, eixo_y):
    plt.figure(figsize=(10,10))
    plt.title(titulo)
    plt.xlabel(eixo_x)
    plt.ylabel(eixo_y)

def plota_set(nome, data):
    prepara_figura(nome, 'tempo (s)', 'aceleração (m/s^2)')
    plt.plot(data.time, data.TgF)
    
    
#Ignora o comeco do movimento, ve a primeira queda
def find_init(data):
    init = 0
    while data.TgF[init] < 2.9:
        init+=1
    return init

#Ignora o fim, ve o ultimo momento que o G e signifcantemente
# menor que 1
def find_end(data):
    end = data.TgF.size-1
    while data.TgF[end] > 0.9:
        end -= 1
    return end
    
    
#FUNCAO PARA ACHAR O PERIODO DO PENDULO    
def gets_period(data, period_array):    
    init = find_init(data)
    print(init)
    term = find_end(data)
    print(term)
    period = 0
    its = 1
    start = 0
    end = 0
    top = 0
    for i in range(init, term):
        if(data.TgF[i] > data.TgF[i+1] and data.TgF[i+1] < data.TgF[i+2]):
            if top == 0:
                start = data.time[i+1]
            elif top == 1: 
                end = data.time[i+1]
        elif(data.TgF[i] < data.TgF[i+1] and data.TgF[i+1] > data.TgF[i+2]):    
            if start != 0 and end==0:
                top+=1
            elif start!=0 and end!=0:
                top +=1
        if top==2 and start!= 0 and end != 0:
            period_array.append(period)
            period+= end-start
            its+=1
            start = 0
            end = 0
            top = 0
    period = period/its  
    return period


```

```python
p1 = pandas.read_csv('data/pendulo/pendulo1.csv', sep=',')
period_array=[]
period = gets_period(p1, period_array)
print(period_array)
print("Periodo: {}".format(period))
plota_set("pOne", p1)
```
